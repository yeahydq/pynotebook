```python
df=pd.DataFrame({'key1':['a','a','b','b','a'],
                 'key2':['one','two','one','two','one'],
                 'data1':np.random.randint(0,10,5),
                 'data2':np.random.randint(0,10,5),
                 })
df
```

```python
df.groupby('key1').sum()
```

```python
df.groupby('key1').max()
```

```python
df.groupby('key1').min()
```

```python
df.groupby('key1').describe()
```

```python
grouped=df.groupby('key1')
```

```python
def peek_range(s):
    return s.max()-s.min()

grouped.agg(peek_range)
```

```python
df
```

```python
grouped.agg(['std','mean','sum',('range',peek_range)])
```

```python
d={'data1':['mean',('range',peek_range)],
   'data2':'sum',
   }
grouped.agg(d)
```

```python
grouped.agg(d).reset_index()
```

```python
df=pd.DataFrame(np.random.randint(0,10,(5,5)),
                columns=list('abcde'),
                index=['alice','bob','Candy','dark','Emily'])
df
```

```python
def demean(s):
    return s-s.mean()
key=['one','one','two','one','two']
demeaned=df.groupby(key).transform(demean)
demeaned
```

```python
demeaned.groupby(key).mean()
```

```python
df=pd.DataFrame({'key1':['a','a','b','b','a','b','a','b','b','a'],
                 'key2':['one','two','one','two','one','two','two','two','one','one'],
                 'data1':np.random.randint(0,10,10),
                 'data2':np.random.randint(0,10,10),
                 }
)
df
```

```python
def top(g,n=2,column='data1'):
    return g.sort_values(by=column,ascending=False)[:n]
top(df)
```

```python
df.groupby('key1').apply(top,3,column='data2')
```

```python
states=['Ohia','New York','Vermont','Floriada',
        'Oregon','Nevada','California','Idaho']
group_key=['East']*4+['West']*4
data=pd.Series(np.random.randn(8),index=states)
data[['Vermont','Nevada','Idaho']]=np.nan
data
```

```python
data.groupby(group_key).mean()
```

```python
data.groupby(group_key).apply(lambda g:g.fillna(g.mean()))
```

```python
fill_mean=lambda g:g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```
