```python
movie_names=['movie_id','title','movie type']
movies=pd.read_table('movies.dat',sep='::',header=None,names=movie_names)
```

```python
rating_names=['user_id','movie_id','rating','time']
ratings=pd.read_table('ratings.dat',sep='::',header=None,names=rating_names)
```

```python
movies.head(3)
```

```python
ratings.head(3)
```

```python
data=pd.merge(pd.merge(users,ratings),movies)
```

```python
print len(data)
data.head(10)
```

```python
data[data.user_id==1][:10]
```

```python
ratings_by_gender=data.pivot_table(values='rating',index='title',columns='gender',aggfunc='mean')
```

```python
ratings_by_gender.head(10)
```

```python
ratings_by_gender['diff']=abs(ratings_by_gender.F-ratings_by_gender.M)
```

```python
ratings_by_gender.head(3)
```

```python
ratings_by_gender.sort_values(by='diff',ascending=False).head(3)
```

```python
ratings_by_title=data.groupby('title')
```
