```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
ts=pd.Series(np.random.randn(1000),index=pd.date_range('2000-01-01',periods=1000))
ts=ts.cumsum()
ts.describe()
```

```python
ts[:10]
```

```python
ts.plot()
```

```python
ts.plot(title='cumsum');
```

```python
df=pd.DataFrame(np.random.randn(1000,4),index=ts.index,columns=list('ABCD'))
df=df.cumsum()
df.describe()
```

```python
df.plot(title='cumsum');
```
